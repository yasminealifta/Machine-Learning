**DATA PREPROCESSING**

Karena pada dataset terdapat beberapa tanda "?" sehinggga dataset tidak bisa diproses lebih lanjut. Agar bisa dilakukan preprocessing maka akan diubah menjadi NaN yang menjadi missing value, jadi dataset bisa di proses dengan mengisi missing value nya nanti.

Terlihat bahwa kolom yang memiliki paling banyak missing value yaitu kolom BMI sebanyak 303 yang merupakan jumlah keseluruhan baris dari dataset. BMI merupakan informasi yang penting karena bisa memberikan kontribusi penting dalam prediksi penyakit. Sehingga perlu menghitung BMI nya dari indikator weight dan length yang sudah ada.

Untuk missing value yang lainnya masih terhitung kecil dengan rata-rata missing value nya kurang dari 10, sehingga akan diatasi dengan mengisinya menggunakan median dan modus dari kolom yang memiliki missing value tersebut.

```python
# mengisi kolom BMI dengan menggunakan kolom weight dan length yang sudah ada
df['BMI'] = df['Weight'] / (df['Length'] / 100) ** 2
```

```python
# Mengatasi missing values
df.fillna(df.median(numeric_only=True), inplace=True)  # Mengisi dengan median untuk kolom numerik
df.fillna(df.mode().iloc[0], inplace=True)  # Mengisi dengan modus untuk kolom kategorikal
```

Untuk missing value dengan data numerik akan diatasi dengan median dari kalom tersebut agar nilai nya tidak bias (menjadi outlier).

Untuk missing value dengan data kategorikal akan diatasi dengan modus dari kolom tersebut agar data yang kosong terisi dengan nilai yang paling umum di kolom tersebut.

```python
df.isnull().sum()
```

**MENGUBAH SEMUA KOLOM MENJADI BERNILAI NUMERIK**

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
# Mengubah semua kolom menjadi bernilai numerik
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le
```

Semua kolom yang ada pada dataset telah diubah menjadi bernilai numerik semua agar bisa dilakukan normalisasi. Karena jika ada kolom yang berisi teks, maka proses normalisasi tidak bisa dilakukan/error.

```python
# menampilkan kembali data yang sudah diubah menjadi nilai numerik semua
display(df.head(20))
```

**MELAKUKAN NORMALISASI PADA MASING-MASING ATRIBUT**

Untuk normalisasi saya menggunakan MInMaxScaller, karena hanya akan mengubah skala data tanpa mengubah bentuk distribusinya dan rentang nilai yang lebih mudah di interpretasikan (0-1).

```python
# Normalisasi data
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# Menampilkan hasil normalisasi
display(df_scaled.head(20))
```

Dataset tersebut sudah dilakukan normalisasi menggunakan MinMaxScaller, sehingga semua nilai berada dalam rentang 0 hingga 1, tanpa mengubah distribusi data aslinya.
